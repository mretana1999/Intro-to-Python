```python
import math

class Shape(object):
    def __init__(self) -> None:
        self.shape = 'shape'

    def __str__(self) -> str:
        return f"I am a {self.shape}"

class Circle(Shape):
    def __init__(self, radius) -> None:
        self.shape = 'circle'
        self.radius = radius
        self.perimeter = 2 * math.pi * radius
        self.area = math.pi * radius * radius

    def __str__(self) -> str:
        return super().__str__() + f" with radius of {self.radius}, perimeter of {self.perimeter:.2f}, and area of {self.area:.2f}"


###################################################################################
################################ Polygon class ####################################
###################################################################################
class Polygon(Shape):
    def __init__(self) -> None:
        self.shape = 'polygon'
        self.side_lengths = ()
        self.num_of_edges = None
        self.perimeter = None
        self.area = None

    def compute_perimeter(self):
        return sum(self.side_lengths)

    def get_number_of_edges(self):
        return len(self.side_lengths)

    def get_area_regular_poly(self):
        return (self.num_of_edges * math.pow(self.side_lengths[0],2)) / (4 * math.tan(math.pi/self.num_of_edges))

    def __str__(self) -> str:
        return super().__str__() + f" with side lengths {self.side_lengths}, perimeter of {self.perimeter:.2f}, and area of {self.area:.2f}"


###################################################################################
################################ Triangle class ###################################
###################################################################################
class Triangle(Polygon):
    def __init__(self, *side_lengths) -> None: #side lengths takes varying positional arguments
        self.shape = 'triangle'
        self.num_of_edges = 3
        self.side_lengths = side_lengths
        if len(self.side_lengths) == 1:
            self.side_lengths = self.side_lengths * 3
        if len(self.side_lengths) != 3:
            raise ValueError('The provided Side Lengths do not match number of edges (sides)')
        if not self.is_valid():
            print(f"Error: Your triangle with side lengths {self.side_lengths} is NOT valid!")
            self.perimeter = None
            self.area = None
        else:
            self.perimeter = self.compute_perimeter()
            self.area = self.get_area()
    
    def get_area(self):
        s = self.compute_perimeter() / 2
        area = s
        for i in range(3):
            area *= s - self.side_lengths[i]
        return area ** 0.5
    
    def is_valid(self):
        a,b,c = self.side_lengths
        return a+b>=c and b+c>=a and c+a>=b

###################################################################################
################################ Rectangle class ##################################
###################################################################################
class Rectangle(Polygon):
    def __init__(self, *side_lengths) -> None:
        self.shape = 'rectangle'
        self.num_of_edges = 4
        self.side_lengths = side_lengths
        if len(self.side_lengths) == 1:
            self.side_lengths = self.side_lengths * 4
        elif len(self.side_lengths) == 2:
            self.side_lengths = self.side_lengths * 2
        if len(self.side_lengths) != 4:
            raise ValueError('The provided Side Lengths do not match number of edges (sides)')
        self.perimeter = self.compute_perimeter()
        self.area = self.get_area()

    def get_area(self):
        return self.side_lengths[0] * self.side_lengths[1]

###################################################################################
################################ Pentagon class ###################################
###################################################################################
class Pentagon(Polygon):
    def __init__(self, *side_lengths) -> None:
        self.shape = 'Pentagon'
        self.num_of_edges = 5
        self.side_lengths = side_lengths
        if len(self.side_lengths) == 1:
            self.side_lengths = self.side_lengths * 5
        if len(self.side_lengths) != 5:
            raise ValueError('The provided Side Lengths do not match number of edges (sides)')
        self.perimeter = self.compute_perimeter()
        self.area = self.get_area_regular_poly()

        #self.area = self.get_area()
    #def get_area(self):
       # return 1/4*(math.sqrt(5*(5+2*math.sqrt(5)))*((self.side_lengths[0])**2))

###################################################################################
################################ Hexagon class ####################################
###################################################################################
class Hexagon(Polygon):
    def __init__(self, *side_lengths) -> None:
        self.shape = 'Hexagon'
        self.num_of_edges = 6
        self.side_lengths = side_lengths
        if len(self.side_lengths) == 1:
            self.side_lengths = self.side_lengths * 6
        if len(self.side_lengths) != 6:
            raise ValueError('The provided Side Lengths do not match number of edges (sides)')
        self.perimeter = self.compute_perimeter()
        self.area = self.get_area_regular_poly()

        #self.area = self.get_area()
    #def get_area(self):
        #return (1/2)*(3*math.sqrt(3))*((self.side_lengths[0])**2)

###################################################################################
################################  Main Function  ##################################
###################################################################################
print("###################################################################################")
print("###################################################################################")
print('''
Welcome to Assignment #8: 
The purpose of this assignment is to design and implement derived [inherited]
classes for the shapes of a circle, triangle, rectangle, pentagon, and a hexagon
''')
print("###################################################################################")
print("###################################################################################")

# Create variables
# Create circle
my_circle = Circle(2)
print(my_circle)

# Create a wrong triangle, will produce error
# my_triangle = Triangle(3, 1.7, 4.9)

# Create a triengle passing the length of each side
my_triangle = Triangle(3, 7, 4.6) 
print(my_triangle)

# attempt to create a triangle with "incorrect" values# should produce error message
my_wrong_triangle = Triangle(3, 1.7, 4.9)

# Create a rectangle of sides 3 and 4.5
my_rectangle = Rectangle(3, 4.5)
print(my_rectangle)

# Create a Pentagon with sides of equal length
my_pentagon = Pentagon(3)
print(my_pentagon)

# Create a Hexagon with sides of equal length
my_hexagon = Hexagon(3)
print(my_hexagon)

########### Print are and prerimeter for each variable ########################
print("###################################################################################")
print("###################################################################################")
print("Circle :")
print(my_circle.area)
print(my_circle.perimeter)
print("")

print("Triangle :")
print(my_triangle.area)
print(my_triangle.perimeter)
print("")

print("Triangle (invalid) :")
print(my_wrong_triangle.area)
print(my_wrong_triangle.perimeter)
print("")

print("Rectangle :")
print(my_rectangle.area)
print(my_rectangle.perimeter)
print("")

print("Pentagon :")
print(my_pentagon.area)
print(my_pentagon.perimeter)
print("")

print("Hexagon :")
print(my_hexagon.area)
print(my_hexagon.perimeter)
print("")
print("###################################################################################")
print("###################################################################################")

```
